```python
ArabicWE = fasttext.load_model('/content/cc.ar.300.bin')
```

```python
WE = [i for i in ArabicWE['سيارة']]
len(WE)
```

```python
def ArabicWEFunction(Word) : 
    WE = [i for i in ArabicWE[Word]]
    return WE      
```

```python
ArabicWEFunction('سيارة')
```

```python
ArabicWEFunction('كتاب')
```

```python
def SimilarWords(Word) :
   
    if Word in ArabicWE.get_words() : 
        WordsDifference = {}
        for EachWord in ArabicWE.get_words()[:10000] : 
            WordsDifference[EachWord] = sum([abs(float(i)-float(j)) for i,j in zip([i for i in ArabicWE[Word]],[i for i in ArabicWE[EachWord]])])
        return list({k: v for k, v in sorted(WordsDifference.items(), key=lambda item: item[1])}.items())[:10]
    else : 
        return 'Word not exists'
```

```python
SimilarWords('سيارة')
```

```python
SimilarWords('يلعب')
```
